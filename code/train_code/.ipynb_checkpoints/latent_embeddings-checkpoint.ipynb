```python
'''
    Plot the following:
    
    Gaussians
    Target embedding before adaptation
    Target embedding after adaptation
'''
```

```python
import time

import matplotlib.pyplot as plt

import numpy as np
import os
from tqdm import tqdm

import config as config
import metrics as metrics

from trainer import MultiSourceTrainer
import gaussian_utils

from tqdm import tqdm

import torch

import umap
```

```python
import config as config
print(config.dataset_name)
print(config.data_key)
```

```python
# Load a source trainer
if config.dataset_name == 'office-caltech':
    src_domain_idx = 2
elif config.dataset_name == 'image-clef':
    src_domain_idx = 0
src_domain_idx = 2
    
print(config.settings['src_datasets'][src_domain_idx])

trainer_S = MultiSourceTrainer(src_domain_idx)
```

```python
trainer_S = MultiSourceTrainer(src_domain_idx)
trainer_S.load_model_weights(it_thresh='enough_iter')
trainer_S.val_over_target_set(save_weights=False)

trainer_S.load_model_weights(it_thresh='max_iter')
trainer_S.val_over_target_set(save_weights=False)
```

```python
# Get the source-only target embedding
trainer_S.load_model_weights(it_thresh='enough_iter')
trainer_S.get_all_val_target_dataloaders()

z_so = []
y_so = []

target_dl_iter_val_list = trainer_S.target_dl_iter_val_list

for data in tqdm(target_dl_iter_val_list):
    indx,images,label,_         = data

    x                           = images.to(trainer_S.settings['device']).float()
    label                       = label.to(trainer_S.settings['device']).long()
    
    with torch.no_grad():
        z_so.append(trainer_S.network.model['global']['Fs'](x).detach().cpu().numpy())
        y_so.append(label.detach().cpu().numpy())
        
z_so = np.vstack(z_so)
y_so = np.concatenate(y_so)
```

```python
trainer_S.load_model_weights(it_thresh='enough_iter')
gaussian_utils.learn_gaussians(trainer_S, debug=True)

# Sample from the gaussian distribution
z_g, y_g = gaussian_utils.sample_from_gaussians(trainer_S.means, trainer_S.covs, n_samples=np.unique(y_so, return_counts=True)[1])
```

```python
# Get the adapted target embedding
trainer_S.load_model_weights(it_thresh='max_iter')
trainer_S.get_all_val_target_dataloaders()

z_a = []
y_a = []

target_dl_iter_val_list = trainer_S.target_dl_iter_val_list

for data in tqdm(target_dl_iter_val_list):
    indx,images,label,_         = data

    x                           = images.to(trainer_S.settings['device']).float()
    label                       = label.to(trainer_S.settings['device']).long()
    
    with torch.no_grad():
        z_a.append(trainer_S.network.model['global']['Fs'](x).detach().cpu().numpy())
        y_a.append(label.detach().cpu().numpy())
        
z_a = np.vstack(z_a)
y_a = np.concatenate(y_a)
```

```python
# Sample from the gaussian distribution and compute latent embeddings
start_time = time.time()

X = np.vstack([z_so, z_a, z_g])
Y = np.concatenate([y_so, y_a, y_g])

reducer = umap.UMAP()
umap_embedding = reducer.fit_transform(X)

print("Took", time.time() - start_time)
```

```python
plt.rcParams['figure.figsize']=(10,8)
my_s = 1
```

```python
# Plot the embedding
l1 = 0; r1 = z_so.shape[0]
l2 = r1; r2 = z_so.shape[0] + z_a.shape[0]
l3 = r2; r3 = X.shape[0]

plt.xticks([], [])
plt.yticks([], [])
plt.scatter(umap_embedding[l1:r1,0], umap_embedding[l1:r1,1], color='teal', label='source-only', s = my_s)
plt.scatter(umap_embedding[l3:r3,0], umap_embedding[l3:r3,1], color='red', label='gaussians', s = my_s)
plt.show()

# l = z_so.shape[0]; r = z_so.shape[0] + z_a.shape[0]
plt.xticks([], [])
plt.yticks([], [])
plt.scatter(umap_embedding[l3:r3,0], umap_embedding[l3:r3,1], color='red', label='gaussians', s = my_s)
plt.scatter(umap_embedding[l2:r2,0], umap_embedding[l2:r2,1], color='orange', label='adapted', s = my_s)
plt.show()

# l = z_so.shape[0] + z_a.shape[0]; r = X.shape[0]
plt.xticks([], [])
plt.yticks([], [])
plt.scatter(umap_embedding[l3:r3,0], umap_embedding[l3:r3,1], color='red', label='gaussians', s = my_s)
plt.scatter(umap_embedding[l1:r1,0], umap_embedding[l1:r1,1], color='teal', label='source-only', s = my_s)
plt.scatter(umap_embedding[l2:r2,0], umap_embedding[l2:r2,1], color='orange', label='adapted', s = my_s)
plt.savefig("../../figures/latent_embedding_{}_{}_src={}".format(config.dataset_name, config.data_key, config.settings['src_datasets'][src_domain_idx]),
           bbox_inches = 'tight')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), markerscale=10., fontsize=26)
plt.savefig("../../figures/latent_embedding_{}_{}_src={}_with_legend".format(config.dataset_name, config.data_key, config.settings['src_datasets'][src_domain_idx]),
           bbox_inches = 'tight')
plt.show()
```
